## 25. Find the maximum time of song listening per day.

```python
df40 = df3.groupby('Date', as_index=False)['msPlayed'].sum()

#df40 = df39.set_index('Date')

df41 = df40.sort_values(by="msPlayed", ascending=False)

df41
```

```python
#function for converting msplayed to played_in_min_sec

def func(msPlayed):
    msPlayed = int(msPlayed)
    m = (msPlayed/1000)/60
    return f"{int(m)} min {round((m%1)*60)} sec"
```

```python
df41.dtypes
```

```python
df41['played_in_m_s'] = df41['msPlayed'].apply(func)

df41
```

```python
df42 = df41.set_index('Date')

df42
```

```python
df42.plot()

plt.xlabel("Date")
plt.ylabel("played_in_m_s")
plt.xticks(rotation=90)
plt.title("Total Maximum Duration(In Descending Order)")
```
